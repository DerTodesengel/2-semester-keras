```python
import random
import os
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from keras import datasets, layers, models, optimizers
import numpy as np
from PIL import Image
from matplotlib import image
import math
imageNum = 1600
train_data = []
train_labels = []
test_data = []
test_labels = []
random.seed(1)
counter = 0
for dirname, _, filenames in os.walk(&#39;input&#39;):
    if len(filenames) == 0:
        continue
    print(dirname)
    filesrnd = random.choices(filenames[:imageNum + 400], k=imageNum + 400)
    randfilenames = filesrnd[:imageNum]
    randfilenames2 = filesrnd[imageNum:imageNum + 400]
for file in randfilenames:
    img = Image.open(os.path.join(dirname, file)).convert(&#39;RGB&#39;)
    img = img.resize((64, 64), image.NEAREST)
    img = np.array(img)
    img = img / 255.0
    train_data.append(img)
    train_labels.append(counter)
for file in randfilenames2:
    img = Image.open(os.path.join(dirname, file)).convert(&#39;RGB&#39;)
    img = img.resize((64, 64), image.NEAREST)
    img = np.array(img)
    img = img / 255.0
    test_data.append(img)
    test_labels.append(counter)
    counter += 1
model = models.Sequential()
```

```python
model.add(layers.Conv2D(32, (3, 3), activation=&#39;relu&#39;, input_shape=(64, 64,
3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation=&#39;relu&#39;))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation=&#39;relu&#39;))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation=&#39;relu&#39;))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(64, activation=&#39;relu&#39;))
model.add(layers.Dense(6, activation=&quot;softmax&quot;))
model.summary()
model.compile(optimizer=optimizers.Adam(learning_rate=0.001),
loss=tf.keras.losses.SparseCategoricalCrossentropy(),
metrics=[&#39;accuracy&#39;])
history = model.fit(np.array(train_data), np.array(train_labels),
batch_size=32, epochs=50,
validation_data=(np.array(test_data),
np.array(test_labels)))
plt.plot(history.history[&#39;accuracy&#39;], label=&#39;accuracy&#39;)
plt.plot(history.history[&#39;val_accuracy&#39;], label=&#39;val_accuracy&#39;)
plt.xlabel(&#39;Epoch&#39;)
plt.ylabel(&#39;Accuracy&#39;)
plt.ylim([0, 1])
plt.legend(loc=&#39;lower right&#39;)
plt.show()
test_loss, test_acc = model.evaluate(np.array(test_data),
np.array(test_labels))
model.summary()
print(test_acc)
```
